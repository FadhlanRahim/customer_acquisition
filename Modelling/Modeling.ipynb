```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfbackup = pd.read_csv('new_df.csv')
df=dfbackup.copy()
df.head()
```

```python
X = df.drop('enrolled', axis=1)
y = df['enrolled']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, stratify = y, random_state=42)
```

```python
train_id=X_train['user']
X_train.drop('user', axis=1, inplace=True)
test_id=X_test['user']
X_test.drop('user', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train))
X_test_scaled = pd.DataFrame(scaler.transform(X_test))
```

```python
X_train_scaled.columns = X_train.columns.values
X_test_scaled.columns = X_test.columns.values
```

```python
X_train_scaled.head()
```

```python
X_train_scaled.index = X_train.index.values
X_test_scaled.index = X_test.index.values
```

```python
X_train_scaled.head()
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state = 0)
clf.fit(X_train_scaled, y_train)
```

```python
y_pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvscore = cross_val_score(clf, X_train_scaled, y_train, cv=10)
```

```python
print(cvscore)
```

* Acuracy = 77%

```python
final_result = pd.concat([test_id, y_test], axis=1)
```
